#Groupby

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df=sns.load_dataset("planets")
df.head()
```

```python
df.describe()
```

```python
df.groupby("method")[["distance"]].agg(['count','min','max','mean', 'std'])
```

```python
df.groupby("method").mean()
```

```python
%matplotlib inline
```

```python
df.groupby("method")[["distance"]].agg(['count','min','max','mean', 'std']).plot(kind='bar')
```

```python
df.groupby("method").agg(['count','min','max','mean', 'std'])
```
